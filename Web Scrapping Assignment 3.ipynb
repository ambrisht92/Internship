Q4- Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com")
```

```python
input1=driver.find_element(By.XPATH,'//input[@class="Pke_EE"]')
input1.send_keys('Oneplus Nord')
```

```python
search1=driver.find_element(By.XPATH,'//button[@class="_2iLD__"]')
search1.click()
```

```python
brandname=[]
brnd=driver.find_elements(By.XPATH,'//div[@class="_4rR01T"]')
for i in brnd:
    brandname.append(i.text)
    
ram_rom1=[]
ram=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][1]')
for r in ram:
    ram_rom1.append(r.text)

disp_size1=[]
disp=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][2]')
for d in disp:
    disp_size1.append(d.text)

camera1=[]   
cam=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][3]')
for camm in cam:
    camera1.append(camm.text)

battery1=[]
bat=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][4]')
for ba in bat:
    battery1.append(ba.text)

warrenty1=[]
waren=driver.find_elements(By.XPATH,'//li[@class="rgWa7D"][5]')
for war in waren:
    warrenty1.append(war.text)
    
price11=[]
pri=driver.find_elements(By.XPATH,'//div[@class="_30jeq3 _1_WHN1"]')
for pric in pri:
    price11.append(pric.text)

pro_url=[]
pro=driver.find_elements(By.XPATH,'//a[@class="_1fQZEK"]')
for url1 in pro:
    pro_url.append(url1.get_attribute('href'))

    
ram_list = []
rom_list = []
for description in ram_rom1:
    parts = description.split('|')
    if len(parts) == 2:
        ram, rom = map(str.strip, parts)
        ram_list.append(ram)
        rom_list.append(rom)

flipkart_product=pd.DataFrame({'Brand Name':brandname,'Price':price11,'Ram':ram_list,'Rom':rom_list,'Display Size':disp_size1,'Camera':camera1,'Battery Size':battery1,'Warrenty':warrenty1,'Product URL':pro_url})
flipkart_product
```

Q6: Write a program to scrap all the available details of best gaming laptops from digit.in.

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.digit.in/top-products/best-gaming-laptops-40.html')
```

```python
lap_name1=[]
lap=driver.find_elements(By.XPATH,'//span[@class="datahreflink"]')
for i in lap:
    lap_name1.append(i.text)

from selenium.common.exceptions import NoSuchElementException
processor=[]
try:
    proc=driver.find_elements(By.XPATH,'//*[@class="Spcs-details"]/table/tbody/tr[2]/td[3]')
    for p in proc:
        processor.append(p.text)
except NoSuchElementException:
    processor.append('-')

dis=[]
disp=driver.find_elements(By.XPATH,'//*[@class="Spcs-details"]/table/tbody/tr[3]/td[3]')
for d in disp:
    dis.append(d.text)

os=[]
os_name=driver.find_elements(By.XPATH,'//*[@class="Spcs-details"]/table/tbody/tr[4]/td[3]')
for o in os_name:
    os.append(o.text)
    
memory=[]
mem=driver.find_elements(By.XPATH,'//*[@class="Spcs-details"]/table/tbody/tr[5]/td[3]')
for i in mem:
    memory.append(i.text)

graphics=[]
grap=driver.find_elements(By.XPATH,'//*[@class="Spcs-details"]/table/tbody/tr[6]/td[3]')
for g in grap:
    graphics.append(g.text)
    
body_weight=[]
body=driver.find_elements(By.XPATH,'//*[@class="Spcs-details"]/table/tbody/tr[7]/td[3]')
for b in body:
    body_weight.append(b.text)
    
pri=[]
pric=driver.find_elements(By.XPATH,'//*[@class="Spcs-details"]/table/tbody/tr[8]/td[3]')
for pr in pric:
    pri.append(pr.text)
    
laptop_details=pd.DataFrame({'Laptop Name':lap_name1[:7],'Processor':processor,'Display':dis,'OS name':os,'Memory':memory,'Graphics':graphics,'Body weight':body_weight,'Price':pri})
laptop_details
```
