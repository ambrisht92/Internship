```python
#Write a python program to display all the header tags from wikipedia.org and make data frame.

import requests
from bs4 import BeautifulSoup
page=requests.get('https://www.wikipedia.org/')
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content)
soup.div
```

```python
#for head in soup.find_all('h1',class_='class="central-textlogo-wrapper'):
#    h.append(head.text)

soup.find_all('h2',class_='central-textlogo-wrapper')
soup.text
```

```python
h1=[]
for i in soup.find('h1'):
    h1.append(i.text)
h1
```

```python
import pandas as pd
df=pd.DataFrame(h1)
df
```

```python
df.drop_duplicates()
```

```python
#2) Write s python program to display list of respected former presidents of India(i.e. Name , Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame.
  
page2=requests.get('https://presidentofindia.nic.in/former-presidents.html')
page2.content
    
```

```python
soup=BeautifulSoup(page2.content)
name=[]
for i in soup.find('div',class_="wgl-services_content-wrap"):
    name.append(i.text)
name
```
