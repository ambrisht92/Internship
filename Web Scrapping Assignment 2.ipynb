```python
# Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. 
#You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
import requests
import pandas as pd
from bs4 import BeautifulSoup
page1=requests.get('https://www.shine.com/job-search/data-analyst-jobs-in-bangalore?q=Data%20Analyst&loc=Bangalore')
a1=page1.content
soup1=BeautifulSoup(a1)
job_title1=[]
location1=[]
exp=[]
comp_name=[]
for i in soup1.find_all('h2',itemprop="name"):
    job_title1.append(i.text)
for i in soup1.find_all('div',class_="jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"):
    location1.append(i.text)
for i in soup1.find_all('div',class_="jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"):
    exp.append(i.text)
for i in soup1.find_all('div',class_="jobCard_jobCard_cName__mYnow"):
    comp_name.append(i.text)
df=pd.DataFrame({'Job Title':job_title1[:10],'Experience':exp[:10],'Company Name':comp_name[:10],'Location':location1[:10]})
df
```

```python
#Q3

import requests
import pandas as pd
from bs4 import BeautifulSoup
page=requests.get('https://www.shine.com/job-search/data-scientist-python-machine-learning-scikit-learn-jobs-in-noida-hyderbaad?q=Data%20Scientist,%20Python,%20Machine%20Learning,%20Scikit-learn,%20&loc=Noida,%20Hyderbaad&location=406&fsalary=1')
a=page.content
soup=BeautifulSoup(a)
job_title=[]
experience=[]
location=[]
company_name=[]
for i in soup.find_all('h2',itemprop="name"):
    job_title.append(i.text)
for i in soup.find_all('div',class_="jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"):
    experience.append(i.text)
for i in soup.find_all('div',class_="jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"):
    location.append(i.text)
for i in soup.find_all('div',class_="jobCard_jobCard_cName__mYnow"):
    company_name.append(i.text)
df=pd.DataFrame({'Job Title':job_title[:10],'Experience':experience[:10],'Company Name':company_name[:10],'Location':location[:10]})
df['Location']=df['Location'].replace('Delhi+6Noida, Bangalore, Chennai, Hyderabad, Pune, Mumbai City','Delhi')
df['Location']=df['Location'].replace('Delhi+4Faridabad, Ghaziabad, Noida, Gurugram','Delhi')
df
```

```python
#Q4 Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
#7. Product
#6. Brand
#7. Product

page3=requests.get('https://www.flipkart.com/search?q=sunglasses+men&sid=26x&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=sunglasses+men%7CSunglasses&requestId=7e807a02-5c60-4376-974a-f565bab3f33a&as-searchtext=sung')
soup3=BeautifulSoup(page3.content)
brand_name=[]
description=[]
price=[]
offer=[]
for name in soup3.find_all('div',class_="_2WkVRV"):
    brand_name.append(name.text)
for prod_des in soup3.find_all('a',class_="IRpwTa"):
    description.append(prod_des.text)
for p in soup3.find_all('div',class_="_30jeq3"):
    price.append(p.text)
for o in soup3.find_all('div',class_="_3Ay6Sb"):
    offer.append(o.text)
page4=requests.get('https://www.flipkart.com/search?q=sunglasses+men&sid=26x&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=sunglasses+men%7CSunglasses&requestId=7e807a02-5c60-4376-974a-f565bab3f33a&as-searchtext=sung&page=2')
soup4=BeautifulSoup(page4.content)
brand_name1=[]
description1=[]
price1=[]
offer1=[]
for name1 in soup4.find_all('div',class_="_2WkVRV"):
    brand_name1.append(name1.text)
for prod_des1 in soup4.find_all('a',class_="IRpwTa"):
    description1.append(prod_des1.text)
for p1 in soup4.find_all('div',class_="_30jeq3"):
    price1.append(p1.text)
for o1 in soup4.find_all('div',class_="_3Ay6Sb"):
    offer1.append(o1.text)
brand_name.extend(brand_name1)
description.extend(description1)
price.extend(price1)
offer.extend(offer1)
df=pd.DataFrame({'Product brand Name':brand_name[:80],'Product Description':description[:80],'Product Price':price[:80],'Product Offer':offer[:80]})
df
```
